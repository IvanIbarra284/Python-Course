#Estructuras de datos
Las estructuras de datos son formas de organizar información para trabajar con ellas de manera eficiente.

- Listas: Pueden ser ordenadas, modificadas y permiten duplicados.
- Tuplas: ordenadas, inmutables y permiten duplicados.
- Sets: no ordenadas, no permiten duplicados.
- Diccionarios: pares clave:valor.

Nota: existen otras que se llaman pilas, colas y listas enlazadas/doblemente enlazadas.

##Listas
Las listas son colecciones ordenadas en las que podemos modificar y almacenar elementos de diferente tipo.

```python
lista_de_compras = ["manzana", "pan", "leche", "huevos", "mantequilla", "queso", "jamon"]
print(lista_de_compras)
print(lista_de_compras[0])
```

```python
#Agregar un elemento a la lista
lista_de_compras.append("arroz")
print(lista_de_compras)
```

```python
lista_de_compras.insert(2, "harina")
print(lista_de_compras)
```

```python
lista_de_compras.remove("harina")
print(lista_de_compras)
```

```python
lista_de_compras.pop() #Sin ningun parametro, el atributo pop borra el ultimo elemento
print(lista_de_compras)

value = lista_de_compras.pop(2)
print(lista_de_compras)
print(value)
```

```python
#Podemos recorrer las listas con los ciclos For
for elemento in lista_de_compras:
  print(elemento)
```

```python
#len() obtiene el tamaño de la lista/array
print(len(lista_de_compras))
for i in range(len(lista_de_compras)):
  print(f"{i+1}.- {lista_de_compras[i]}")
```

```python
promedios = [55.6, 97.5, 23.6, 97.5, 72.4, 100]
promedios.sort() #sort ordena ascendentemente tus arreglos
print(promedios)
```

```python
promedios.sort(reverse=True) #sort ordena descendentemente tus arreglos
print(promedios)
```

```python
#podemos ingresar al ultimo elemento de una lista
print(lista_de_compras[-1])
print(promedios[-1])
```

```python
#Max y Min
print(max(promedios))
print(min(promedios))
```

```python
#Declaración y uso de matrices
calificaciones = [[10, 9, 8], [7, 6, 5]]
print(calificaciones)
i=1
for cal_i in calificaciones:
  j=1
  print(f"Alumno {i}")
  for cal_j in cal_i:
    print(f"Calificacion {j}: {cal_j}")
    j+=1
  i+=1
```

```python
#Array para guardar fechas de nacimiento
fechas = ("10/12/07", "05/02/12", "18/03/05")
print(fechas)
```

```python
#Una tupla es INMUTABLE
fechas.append("18/11/05")
```

```python
fecha_lista = list(fechas)
fecha_lista.append("18/11/05")
print(fecha_lista)
```

##Sets
Colecciones que no permiten duplicados ni tienen orden, pero si pueden guardar multiples tipos de datos a la vez

```python
nombre = {"Angel", "Juan", "Diana", "Eliud", "Juan"}
print(nombre)
```

```python
nombre.add("Leonardo")
print(nombre)
```

```python
#No es posible acceder a valores especificos usando su orden.
print(nombre[2])
```

```python
#Unión de conjuntos
matutino = {20322, 20323, 20324, 20325}
vespertino = {20322, 20, 20328, 23244}
total_matriculas = matutino | vespertino #ó matutino.union(vespertino)
print(total_matriculas)
```

```python
#Intersección de conjuntos
algebra = {20322, 20323, 56324, 74325}
fisica = {20322, 20323, 20328, 23244}
materias = algebra & fisica #ó algebra.intersection(fisica)
print(materias)
```

```python
#Diferencia
solo_algebra = algebra - fisica #ó algebra.difference(fisica)
print(solo_algebra)
```

```python
nombre = {"Angel", "Juan", "Diana", "Eliud", "Juan"}
#remove, pop
nombre.remove("Juan")
print(nombre)
nombre.pop() #En el caso de los sets, elimina un elemento de manera aleatoria
print(nombre)
```

```python
nombre.clear() #Elimina todo el contenido de la variable, pero no la variable
print(nombre)
```

##Diccionarios
Son colecciones no ordenadas de pares "Clave:Valor"

```python
alumno = {
    #Clave  :  Valor
    "Nombre": "Iván Alexis",
    "Edad": 20,
    "Carrera": "ITS",
    "Semestre": 7,
    "Promedio": 80,
    "Calificaciones": [100, 94, 94, 97, 100]
}
print(type(alumno))
alumno
```

```python
print("El alumno", alumno["Nombre"], "tiene un promedio de", alumno["Promedio"], "esta estudiando la carrera:", alumno["Carrera"])
```

```python
#Agregar al alumno...
alumno["Area preferida"] = "Ciencia de datos"
alumno
```

```python
for valor in alumno:
  print(alumno[valor])
```

```python
for clave, valor in alumno.items():
  print(clave,":", valor)
```

```python
#El metodo clear tambien se puede usar en los diccionarios.
#La función get nos permite obtener el valor de cierta clave.
print(alumno.get("Nombre"))
```

```python
alumno_1 = alumno.items()
print(alumno_1)
print(type(alumno_1))
```

```python
alumno_lista = list(alumno_1)
print(alumno_lista)
print(type(alumno_lista))
```

```python
#El método values sirve para obtener los valores de un diccionario. Se puede
alumno_values = alumno.values()
print(alumno_values)
print(type(alumno_values))
```

```python
if "Carrera" in alumno:
  print("Si existe")
else:
  print("No existe")
if "Dinero" in alumno:
  print("Tiene dinero")
else:
  print("No tiene dinero")
```

```python
#copy sirve para copiar todo el diccionario en una nueva variable
#update sirve para actualizar la clave y valor de un diccionario
alumno.update({"Semestre": 5})
alumno
```

```python
# Lista de productos
productos = ["manzana", "banana", "leche", "pan"]  # <-- LISTA

# Tupla de precios
precios = (10, 5, 20, 15)  # <-- TUPLA

# Diccionario de inventario: producto -> {"precio": x, "stock": y}
inventario = {productos[i]: {"precio": precios[i], "stock": 10} for i in range(len(productos))}
# <-- DICCIONARIO
# Clave: nombre del producto
# Valor: otro diccionario con "precio" y "stock"

# Set de productos en oferta
ofertas = {"manzana", "pan"}  # <-- SET

# Mostrar productos
for p in productos:
    msg = "OFERTA!!!" if p in ofertas else "Grandes precios!"
    info = inventario[p]
    print(f"{p.title()} - ${info['precio']} - Stock: {info['stock']} -> {msg}")

# Compra simple
compra = input("\n¿Qué producto quieres comprar? ").strip().lower()

if compra in inventario and inventario[compra]["stock"] > 0:
    inventario[compra]["stock"] -= 1
    print(f"Compraste 1 {compra}. Stock restante: {inventario[compra]['stock']}")
else:
    print("Producto no disponible.")
```
