##Tipos de Datos

```python
#Existen los datos:
enteros = 1 #Son numeros sin decimales
flotantes = 1.1 #Son numeros con decimales
type(enteros)
```

```python
#Cadenas de datos/Strings
caracter = "a"
cadena_de_caracteres = "Hola"
type(cadena_de_caracteres)
```

```python
#Booleanos
verdadero = True
falso = False
type(verdadero)
```

```python
#Imprimir valores
print(enteros)
print(flotantes)
print(caracter)
print(cadena_de_caracteres)
print(verdadero)
print(falso)
```

```python
#Ingresar valores
nombre = input("Ingrese su nombre: ") #La función input permite imprimir valores también, para dar contexto de lo que se necesita ingresar
print("Hola", nombre)

num = int(input("Ingrese un número: ")) #La función int permite guardar el input como un Entero (si es posible) en lugar de un String
print(num + 1)
```

##Operaciones Matemáticas

```python
#4to Nivel de importancia
#Suma Ordinaria
print(3 + 7)
variable1, variable2 = 10, 20
print(variable1 + variable2)

#Resta Ordinaria
print(10 - 5)
print(variable2 - variable1)
```

```python
#3er Nivel de Importancia
#Multiplicación Ordinaria
print(variable1, variable2)
print("------------------------")
print(3 * 7)
print(variable1 * variable2)

#División Ordinaria
print(15 / 5)
print(variable2 / variable1)

#División de piso (Redondea siempre hacia abajo)
print((variable2 + 9) // variable1)

print(variable2 / variable1 * 5)
```

```python
#Potencias
print(variable1, variable2)
print("------------------------")
print(2 ** 3)
print(variable1 ** variable2)
```

##Operaciones Lógicas

```python
#Empezamos con los operadores lógicos <, >, <=, >=, and, or (Siempre retornan un valor)
b1 = 4 > 5
b2 = 7 < 3
b3 = 14 >= 14
```

```python
#Quieres un sandwich...
pan = True
queso = True
mayonesa = True
jamon = False

print(pan and queso and mayonesa or jamon)
print(True * 2)
print(False * 74568)
```
