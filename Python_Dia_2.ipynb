##Elif
Para este tema, realizaremos un ejercicio: un programa que evalue la calificación con las siguientes normas
- A+ = 100
- A = 90-99
- B = 80-89
- C = 70-79
- D = 60-69
- F = 0-59

```python
cal = 80
if cal == 100:
  print("A+")
elif cal >= 90:
  print("A")
elif cal >= 80:
  print("B")
elif cal >= 70:
  print("C")
elif cal >= 60:
  print("D")
else:
  print("F")
```

##Ciclo While

```python
suma = 0
numeros = int(input("Ingresa el numero a sumar (0 = Salir): "))
while numeros != 0:
  suma += numeros
  numeros = int(input("Ingresa el numero a sumar (0 = Salir): "))
print(f"La suma es {suma}")
```

```python
#Sistema que calcula el promedio de un alumno al pedir varias materias, con while
suma, i = 0, 0
materias = int(input("Total de materias a evaluar: "))
print("------------------------")
while i < materias:
  nota = float(input("Ingresa la nota: "))
  if nota >= 0 and nota <= 100:
    suma += nota
    i += 1
  else:
    print("Nota no válida")
print(f"El promedio es {suma/materias}")
```

```python
#Programa que juega "piedra, papel o tijeras" con el usuario
import random

opciones = ["piedra", "papel", "tijeras"]

while True:
    # Pedir la elección del jugador
    eleccion = input("\nElige piedra, papel o tijeras (o escribe salir si ya no quieres jugar): ")

    # Validar la entrada
    if eleccion not in opciones and eleccion != "salir":
        print("Opción no válida, intenta de nuevo.")
        continue  # vuelve a pedir la opción

    # Salir del juego
    if eleccion == "salir":
        print("¡Gracias por jugar!")
        break

    # Elección aleatoria de la computadora
    aleatoria = random.choice(opciones)
    print("La computadora eligió:", aleatoria)

    # Determinar el ganador
    if eleccion == aleatoria:
        print("Empate")
    elif (eleccion == "piedra" and aleatoria == "tijeras") \
         or (eleccion == "papel" and aleatoria == "piedra") \
         or (eleccion == "tijeras" and aleatoria == "papel"):
        print("¡Ganaste! :))")
    else:
        print("Perdiste :((")
```
