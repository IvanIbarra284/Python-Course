##¿Qué es un módulo?

Un modulo es una unidad de organizacion del codigo que permite encapsular y dividir el software en partes mas manejables,
generalmente representado como un archivo que contiene codigo relacionado, como funciones, variables o metodos de clase.

**Existe solo un tema que TODO programador independientemente de su área debe saber:**

JSON es un formato para guardar y transmitir datos. Es muy común en el uso de API's, archivos de configuración y almacenamiento de información.
Se parece a un diccionario, pero solo contiene texto.

Sirve para convertir objetos de python -> texto JSON y también texto JSON -> objetos de python. A esto se le llama Serializar (DE objeto de python a JSON) y Deserializar (De JSON a objetos de python).

```python
#Funcioness principales
# .json.dums(obj): Convierte un diccionario u objeto de python a texto JSON
import json
obj = {"Nombre":"Ana", "Edad":25}
print(type(obj))
s = json.dumps(obj, ensure_ascii= False)
print(s)
print(type(s))
#
```

```python
# .json.load(texto_json): Convierte un texto JSON a diccionario u objeto de python
restaurar = json.loads(s)
print(restaurar)
print(type(restaurar))
```

###Librería "numpy"
Provee el tipo ndarray, operaciones vectorizadas, brodcasting, utilidades de algebra lineal y
funciones matematicas en arrays. Es la base sobre la que se apoyan muchas librerias cientificas en Python

**NOTA** El arreglo de numpy no es igual a la lista de pyhton, puesto que no puede tener diferentes tipos de datos.

```python
import numpy as np
A = np.array([[1,2,3],[4,5,6]]) #array (2,3)
V = np.array([10,20,30]) #array (3)
print(A)
print(V)
print("------------------")
print(A+V)
print("------------------")
prueba1 = [1, 2, 3]
prueba2 = [1, 2, 3]
print(prueba1+prueba2)
```

```python
z = np.zeros((3,3)) #array (3,3) lleno de ceros
print(z)
print("------------------")
o = np.ones((2,4)) #array (3,3) lleno de ceros
print(o)
print("------------------")
r = np.random.rand(3,3) #array (2,2) lleno de numeros aleatorios entre 0 y 1
print(r)
```

```python
#Rangos numéricos con numpy
nums = np.arange(0,10,2) #saca los valores entre 0 y 10, de 2 en dos
print(nums)
x = np.linspace(0,1,5) #saca 5 valores entre 0 y 1, espaciados de manera equitativa
print(x)
```

```python
a = np.array([1,2,3,4,5])
b = np.array([6,7,8,9,10])
#suma, resta, multiplicación division y potencia de matrices
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a**b)
```

```python
print(a.dot(b)) #imprime la multiplicacion de la matriz a por b
c = np.array([[1,2],[3,4],[5,6]])
d = np.array([[7,8],[9,10],[11,12]])

# Para la multiplicación de matrices, el número de columnas de la primera matriz debe ser igual al número de filas de la segunda.
# c tiene forma (3,2) y d tiene forma (3,2).
# Para corregir el error, transponemos 'd' para que tenga forma (2,3).
print(c.dot(d.T))
print(np.dot(c, d.T))
#
```

###Librería MatplotLib
