##Funciones que llaman a otras funciones

```python
def saludar(nombre):
  return f"Hola {nombre}"

def despedirse(nombre):
  return f"Adios {nombre}"

def saludo(nombre):
  print(saludar(nombre))
  print(despedirse(nombre))

saludo("Victor")
```

```python
#Supongamos que tenemos una heladeria
def pedir_sabor():
  sabores = ["vainilla", "chocolate", "fresa"]
  print("Sabores disponibles",",",sabores)
  #lower() Hace minúsculas las palabras ingresadas
  #strip() Elimina los espacios antes y después de las palabras ingresadas
  sabor = input("¿Qué sabor te gustaría? ").lower().strip()

  while sabor not in sabores:
    print("No tenemos ese sabor")
    sabor = input("¿Qué sabor te gustaría? ").lower().strip()
  return sabor

def pedir_cantidad():
  i=0
  while i==0:
      cantidad = int(input("¿Cuánta desea? "))
      if cantidad > 0:
        i=1
      else:
        print("Debe ser mayor a 0")
  return cantidad

def pedir_helado():
  sabor = pedir_sabor()
  cantidad = pedir_cantidad()
  print("Pedido: ", cantidad,"bola(s) del sabor",sabor)
  return sabor, cantidad

pedir_helado()
```

##Funciones Lambda
nombreFuncion = Lambda parametros: acción

*NOTA* Suelen ser más limitadas que las funciones normales

```python
sumar = lambda x,y : x + y
print(sumar(2,5))
```

```python
descuentoLambda = lambda desc, precio: ((precio*(1-desc)),(precio * descuento))
nuevoPrecio = descuentoLambda(.15, 100)
print("El precio es:",nuevoPrecio[0], "El descuento que se hizo es:", nuevoPrecio[1])
```

```python
#Calculadora (suma, resta, producto, division y potencia)
def suma (a,b):
  return a+b
def resta (a,b):
  return a-b
def producto (a,b):
  return a*b
def division (a,b):
  if b==0:
    print("Math Error")
  return a/b
def potencia (a,b):
  return a**b

def Menu():
  print("-------Calculadora-------")
  print("1. Suma")
  print("2. Resta")
  print("3. Producto")
  print("4. Division")
  print("5. Potencia")
  print("6. Salir")

while True:
  Menu()
  try:
    op = int(input("¿Qué operación deseas realizar? "))
    if op < 1 or op > 6:
      print("Opción no válida")
      continue
    elif op == 6:
      print("Hasta luego")
      break
    else:
      a = float(input("Ingresa el primer número: "))
      b = float(input("Ingresa el segundo número: "))
  except ValueError:
    print("Error: No es un numero")

  try:
    if op == 1:
      print("-------------------------\nLa suma es:",suma(a,b))
    elif op == 2:
      print("-------------------------\nLa resta es:",resta(a,b))
    elif op == 3:
      print("-------------------------\nEl producto es:",producto(a,b))
    elif op == 4:
      print("-------------------------\nLa división es:",division(a,b))
    else:
      print("-------------------------\nLa potencia es:",potencia(a,b))
  except ZeroDivisionError:
      print("Error: No se puede dividir entre 0")
      continue
```
