##Elif
Para este tema, realizaremos un ejercicio: un programa que evalue la calificación con las siguientes normas
- A+ = 100
- A = 90-99
- B = 80-89
- C = 70-79
- D = 60-69
- F = 0-59

```python
cal = 80
if cal == 100:
  print("A+")
elif cal >= 90:
  print("A")
elif cal >= 80:
  print("B")
elif cal >= 70:
  print("C")
elif cal >= 60:
  print("D")
else:
  print("F")
```

##Operadores Lógicos con Condicionales

```python
#Quieres un sandwich...
pan = True
queso = True
mayonesa = True
jamon = False
print (pan and queso and mayonesa and jamon)
if not(pan and queso and mayonesa and jamon):
  print("No puedes tener un sandwich")
```

```python
cal = 80
if cal <= 100 and cal >= 0:
  if cal == 100:
    print("A+")
  elif cal >= 90:
    print("A")
  elif cal >= 80:
    print("B")
  elif cal >= 70:
    print("C")
  elif cal >= 60:
    print("D")
  else:
    print("F")
else:
  print("Calificación no válida")
```

##Ciclo While

```python
suma = 0
numeros = int(input("Ingresa el numero a sumar (0 = Salir): "))
while numeros != 0:
  suma += numeros
  numeros = int(input("Ingresa el numero a sumar (0 = Salir): "))
print(f"La suma es {suma}")
```

```python
#Sistema que calcula el promedio de un alumno al pedir varias materias, con while
suma, i = 0, 0
materias = int(input("Total de materias a evaluar: "))
print("------------------------")
while i < materias:
  nota = float(input("Ingresa la nota: "))
  if nota >= 0 and nota <= 100:
    suma += nota
    i += 1
  else:
    print("Nota no válida")
print(f"El promedio es {suma/materias}")
```
